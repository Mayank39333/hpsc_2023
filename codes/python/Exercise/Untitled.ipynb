```python
import numpy as np
import matplotlib.pyplot as plt
a=np.linspace(0,99,100)
b=[]
def facto(n):
    fact=1
    for i in range(n):
        fact*=i+1
        
    return fact
eps=0.00001
for j in range(0,len(a)):
    e=0
    for i in range(0,100):
        e=e+pow(a[j],i)/facto(i)
        if pow(a[j],i)/facto(i)<eps:
            break
    b.append(e)
# print("converges at "+str(i)+"th iteration")
    
y=np.exp(a)
plt.plot(a,b)
plt.plot(a,y)
%timeit b
%timeit y
```
